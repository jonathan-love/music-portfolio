## Programme Note

Benedictus is a setting of the traditional liturgical text, composed in early 2025 for SATB choir with divisi, tenor solo, and organ. While originally composed as an unaccompanied work, the organ part was later added in response to a suggestion from my supervisor - Phillip Cooke - and as part of my exploration into writing for the instrument for the first time.

This setting aims to offer a fresh perspective on a text that has been set many times before. One of the main ways I approached this was through the use of irregular meter (7/4, subdivided into 2 + 2 + 3). This meter introduces a sense of forward momentum and energy, while the repetition of a central ‚Äúchorus‚Äù motif helps both performers and audience become accustomed to the rhythmic structure. The piece also plays with tonal ambiguity, occasionally resolving to E major at phrase endings to provide moments of clarity amid more modal harmonic uncertainty, taking inspiration from the works of Kenneth Leighton.

A tenor solo section offers textural and harmonic variety, swapping briefly to 4/2. In this passage, the soloist presents the text almost in full, uninterrupted, before the return of the main thematic material. Though the piece is technically demanding and fast-paced, I‚Äôve composed the piece to include breaks and repetitions, giving singers space to breathe and return to familiar material.
